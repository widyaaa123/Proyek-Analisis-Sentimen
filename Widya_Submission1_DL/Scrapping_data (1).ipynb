```python
from youtube_comment_downloader import YoutubeCommentDownloader
import pandas as pd

video_url = "https://youtu.be/1unIUw5HS84?si=_IIOvHHYrYgQ5D__"

# Ambil video ID
video_id = video_url.split('/')[-1].split('?')[0]
print("Video ID:", video_id)

# Inisialisasi downloader
downloader = YoutubeCommentDownloader()

# Ambil komentar (generator)
comments = downloader.get_comments(video_id)

# Simpan ke list
data = []
for c in comments:
    data.append(c)

print("Jumlah komentar yang diambil:", len(data))

# Convert ke DataFrame
df = pd.DataFrame(data)

# Simpan ke CSV
df.to_csv(f"comments_ytb.csv", index=False)

# Lihat 5 baris pertama
df.head()
```
